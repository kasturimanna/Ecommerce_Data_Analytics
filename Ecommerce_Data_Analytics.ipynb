```python
master_df = order_items.merge(orders, on='order_id', how='left')
master_df = master_df.merge(products, on='product_id', how='left')
master_df = master_df.merge(customers, on='customer_id', how='left')
print("Master table shape:", master_df.shape)
print(master_df.head())
```

```python
print(master_df)
```

```python
master_df.columns
```

```python
master_df["revenue"]=master_df["price"]*master_df["quantity"]
```

```python
print(master_df["revenue"])
```

```python
master_df["profit"]=((master_df["price"]-master_df["cost"])*master_df["quantity"])
print(master_df["profit"])
```

```python
print(master_df[["order_id", "product_name", "quantity", "price", "revenue"]].head())
```

```python
customer_features = master_df.groupby('customer_id').agg(
    total_revenue = ('revenue', 'sum'),
    total_orders = ('order_id', 'nunique')  
).reset_index()
customer_features['avg_order_value'] = customer_features['total_revenue'] / customer_features['total_orders']

customer_features = customer_features.merge(
    master_df[['customer_id','name','age','gender','city']].drop_duplicates(),
    on='customer_id',
    how='left'
)

top_customers = customer_features.sort_values(by='total_revenue', ascending=False).head(5)
print(top_customers)
```
